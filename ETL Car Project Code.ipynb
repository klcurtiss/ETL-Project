```python
totalcars = len(data['results'])
totalcars
```

```python
#Define and append lists 
year_list=[]
make_list=[]
category_list=[]
model_list=[]
for car in range(totalcars):
    year_list.append(data['results'][car]['Year'])
    make_list.append(data['results'][car]['Make'])
    model_list.append(data['results'][car]['Model'])
    category_list.append(data['results'][car]['Category'])
    

#print(year_list)
print(make_list)
```

```python
#Create dataframe from lists
SummaryTable = pd.DataFrame({
        "year":year_list,
        "make":make_list,
        "model":model_list,
        'category_list':category_list,
        
})
SummaryTable
```

```python
#merge the two data sources into one dataframe
Auto_Sales = pd.merge(SummaryTable, Auto_Sales_Table, how = "left", on = "make")
Auto_Sales
```
